```python
import numpy as np
from google.colab import autoviz

def heatmap(df, x_colname, y_colname, figscale=1, mpl_palette_name='viridis'):
  from matplotlib import pyplot as plt
  import seaborn as sns
  import pandas as pd
  plt.subplots(figsize=(8 * figscale, 8 * figscale))
  df_2dhist = pd.DataFrame({
      x_label: grp[y_colname].value_counts()
      for x_label, grp in df.groupby(x_colname)
  })
  sns.heatmap(df_2dhist, cmap=mpl_palette_name)
  plt.xlabel(x_colname)
  plt.ylabel(y_colname)
  return autoviz.MplChart.from_current_mpl_state()

chart = heatmap(df, *['Purpose of the trip', 'Type of Stay'], **{})
chart
```

```python
import pandas as pd
import statsmodels.api as sm

# Load your data into a DataFrame
data = pd.read_csv('/content/data30overnt.csv')
print(df.shape[0])
#renaming the coloumns
data.rename(columns={'b4q4': 'Gender', 'b4q5': 'Age',
                   'b4q6':'Marital status', 'b52q5': 'Purpose of the trip'
                   ,'b52q7' :'Mode of travel','b52q9':'Type of Stay',
                   'b52q14':'Main destination' },inplace=True)
print(data.columns)
```

```python
# Map values for the outcome variable (gender)
data['Gender'] = data['Gender'].map({ 2: 'Female',1: 'Male', 3: 'Third Gender'})

# Map values for the dependent variables
data['Marital status'] = data['Marital status'].map({1: 'Never Married', 2: 'Currently Married', 3: 'Widowed', 4: 'Divorced'})
data['Type of Stay'] = data['Type of Stay'].map({1: 'Type of Stay_Hotel', 3: 'Type of Stay_Dharmashala',
                                                 5: 'Type of Stay_Friends/Relatives', 4: 'Type of Stay_Rented House',
                                                 2: 'Type of Stay_Guest House', 9: 'Type of Stay_Others'})
data['Mode of travel'] = data['Mode of travel'].map({'2': 'Mode of Travel_Bus',2: 'Mode of travel_bus', 3: 'Mode of Travel_Train', 6: 'Mode of Travel_Own',
                                                   8: 'Mode of Travel_Motorized Vehicle', 10: 'Mode of Travel_Non-motorized Vehicle',
                                                   '1': 'Mode of Travel_On Foot', 4: 'Mode of Travel_Ship/Boat',
                                                   '1': 'Mode of Travel_Air', '7': 'Mode of Travel_Own Non-Motorized',
                                                   '19': 'Mode of Travel_Others'})
data['Purpose of the trip'] = data['Purpose of the trip'].map({2: 'Purpose of the Trip_Holidaying',
                                                             6: 'Purpose of the Trip_Health/Medical',
                                                             7: 'Purpose of the Trip_Shopping',
                                                             1: 'Purpose of the Trip_Business',
                                                             3: 'Purpose of the Trip_Social',
                                                             4: 'Purpose of the Trip_Pilgrimage & Religious',
                                                             5: 'Purpose of the Trip_Education/Training',
                                                             9: 'Purpose of the Trip_Others'})
data['Main destination'] = data['Main destination'].map({1: 'Destination within the District',
                                                         2: 'Destination outside the District but within the State',
                                                         3: 'Destination outside the District but within the State',
                                                         4: 'Final Part of Departure in Indian Territory for International Trip'})
```

```python
# Define your predictor variables, including the newly mapped ones
predictor_variables = ['Gender', 'Marital status', 'Type of Stay', 'Mode of travel',
                       'Purpose of the trip', 'Main destination', 'Age']

# Create dummy variables for the categorical predictor variables
data = pd.get_dummies(data, columns=['Gender', 'Marital status', 'Type of Stay', 'Mode of travel',
                                     'Purpose of the trip', 'Main destination'], drop_first=True)
```

```python
# Remove leading and trailing spaces from column names
data.columns = data.columns.str.strip()

# Check the column names to verify the corrected name
print(data.columns)

# Access the 'Gender' column
y = data['Gender_Male']
```

```python
predictor_variables = [
       'Marital status_Never Married', 'Marital status_Widowed',
       'Type of Stay_Type of Stay_Friends/Relatives',
       'Type of Stay_Type of Stay_Guest House',
       'Type of Stay_Type of Stay_Hotel', 'Type of Stay_Type of Stay_Others',
       'Type of Stay_Type of Stay_Rented House',
       'Mode of travel_Mode of Travel_Bus',
       'Mode of travel_Mode of Travel_Motorized Vehicle',
       'Mode of travel_Mode of Travel_Non-motorized Vehicle',
       'Mode of travel_Mode of Travel_Others',
       'Mode of travel_Mode of Travel_Own',
       'Mode of travel_Mode of Travel_Own Non-Motorized',
       'Mode of travel_Mode of Travel_Ship/Boat',
       'Mode of travel_Mode of Travel_Train',
       'Purpose of the trip_Purpose of the Trip_Education/Training',
       'Purpose of the trip_Purpose of the Trip_Health/Medical',
       'Purpose of the trip_Purpose of the Trip_Holidaying',
       'Purpose of the trip_Purpose of the Trip_Others',
       'Purpose of the trip_Purpose of the Trip_Pilgrimage & Religious',
       'Purpose of the trip_Purpose of the Trip_Shopping',
       'Purpose of the trip_Purpose of the Trip_Social',
       'Main destination_Destination within the District',
                      'Age']
```

```python

# Access the 'Gender' column
y = data['Gender_Male']
# Add an intercept (constant)
X = sm.add_constant(data[predictor_variables])
```

```python
# Fit a Multinomial Logistic Regression model
mnl_model = sm.MNLogit(y, X)

# Fit the model
mnl_results = mnl_model.fit()

# Print the model summary
print(mnl_results.summary())
```
